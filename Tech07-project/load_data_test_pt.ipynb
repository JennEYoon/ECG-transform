To export a PyTorch array to a CSV file, you can follow these steps:  

Convert the array to a NumPy array:

   import torch
   import numpy as np

   # Create a sample PyTorch tensor
   tensor = torch.randn(3, 2)

   # Convert to NumPy array
   numpy_array = tensor.numpy()

   # Use numpy.savetxt to save the array to CSV:
   np.savetxt("data.csv", numpy_array, delimiter=",")

Alternative method using Pandas:
Convert the array to a Pandas DataFrame:

   import pandas as pd

   # Convert to DataFrame
   df = pd.DataFrame(numpy_array)

   # Use df.to_csv to save the DataFrame to CSV:
   df.to_csv("data.csv", index=False)


```python
import numpy as np
import pandas as pd
import torch
```

```python
ptb_train = torch.load('/content/train.pt')
ptb_test = torch.load('/content/test.pt')
ptb_val = torch.load('/content/val.pt')
```

```python
print(type(ptb_train))
```

```python
# Check each key-value pair in the dictionary
for key, value in ptb_train.items():
    print(f"Key: {key}, Type: {type(value)}")
    if isinstance(value, torch.Tensor):
        print(f"Shape of {key}: {value.shape}")
```

```python
y = ptb_train['labels'].numpy()
X = ptb_train['samples'].numpy()
print(y.shape, X.shape)
```

```python
X = X.squeeze(axis=1)
print(X.shape)
```

```python
df = pd.DataFrame(X)
df
```

```python
df2 = pd.DataFrame(y)
df2
```

```python
# Concatenate df2 as new column to df1
# df_combined = pd.concat([df1, df2], axis=1)
df3 = pd.concat([df, df2], axis=1)
df3
```

```python
# Save DataFrame to .csv without headers
# df.to_csv('output.csv', header=False, index=False)
df3.to_csv('ptb_train.csv', header=False, index=False)
```

```python
y_test = ptb_test['labels'].numpy()
X_test = ptb_test['samples'].numpy()
print(y_test.shape, X_test.shape)
```

```python
X_test = X_test.squeeze(axis=1)
print(X_test.shape)
```

```python
df_test = pd.DataFrame(X_test)
df2_test = pd.DataFrame(y_test)
df3_test = pd.concat([df_test, df2_test], axis=1)
df3_test
```

```python
# Save DataFrame to .csv without headers
# df.to_csv('output.csv', header=False, index=False)
df3_test.to_csv('ptb_test.csv', header=False, index=False)
```

```python
y_val = ptb_val['labels'].numpy()
X_val = ptb_val['samples'].numpy()
X_val = X_val.squeeze(axis=1)
df_val = pd.DataFrame(X_val)
df2_val = pd.DataFrame(y_val)
df3_val = pd.concat([df_val, df2_val], axis=1)
df3_val
```

```python
# Save DataFrame to .csv without headers
df3_val.to_csv('ptb_val.csv', header=False, index=False)
```
