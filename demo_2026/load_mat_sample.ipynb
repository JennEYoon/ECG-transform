### Augmented data sample:  
Verify that sampling rate matches what will be used in inference model run.  
Should be 125 hertz and millivolts amplitude. 

```python
# Load the .mat file
mat_data = scipy.io.loadmat(filename, struct_as_record=False, squeeze_me=True)
print(mat_data, type(mat_data))
```

```python
# Access the struct 
data = mat_data['val']

# Show elements of data (ndarray type)
print('first 5 elements: \n', data[:5])
print('length of row 0:', len(data[0]))
```

```python
# Check units range is +1 to -1.0 (millivolts)
for i in range(12): 
    print(f"min row {i}: {min(data[i])}, max row {i}: {max(data[i])}") 
    next 
# row 7 is slightly beyong -1.0. OK.     
# row labels:(I, II, III, aVR, aVL, aVF, V1, V2, V3, V4, V5, V6) 
```

**AI note: The Formatting Pattern: F-Strings**  
The line in your cell is (or should be) a Formatted String Literal, commonly called an f-string.

The Syntax: It is defined by placing an f or F before the opening quotation mark: f"...".

The Mechanism: Anything placed inside curly braces { } is treated as Python code. Python evaluates that code and converts the result into a string.
