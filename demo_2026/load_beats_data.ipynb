```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data = pd.read_csv("AR2026.03.csv")
```

### Plot some heart beats


```python
### **Date & Time:** Friday, June 19, 2026, 3:48 PM

### Simple Method Plots

# Here is a straightforward way to plot the first 2 and last 2 heartbeats on a single graph:

plt.plot(data.iloc[0], label="First Beat")
plt.plot(data.iloc[1], label="Second Beat")
plt.plot(data.iloc[-2], label="Second-to-Last Beat")
plt.plot(data.iloc[-1], label="Last Beat")
plt.legend()
plt.tight_layout()
plt.savefig("heartbeats_combined.png")
```

```python
### Standard Option (Using Subplots)

# If you prefer to see each heartbeat in its own separate subplot for clearer comparison,
# you can use the following approach:

fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
indices = [0, 1, -2, -1]
labels = ["First Beat", "Second Beat", "Second-to-Last Beat", "Last Beat"]

for ax, idx, label in zip(axes.ravel(), indices, labels):
    ax.plot(data.iloc[idx])
    ax.set_title(label)
    ax.grid(True)

plt.tight_layout()
plt.savefig("heartbeats_subplots.png")
```

```python

### Plot all 4 leads, first person, first beat

# Here is a straightforward way to plot the first 2 and last 2 heartbeats on a single graph:

plt.plot(data.iloc[0], label="I")
plt.plot(data.iloc[1], label="II")
plt.plot(data.iloc[2], label="V2")
plt.plot(data.iloc[3], label="V5")
plt.legend()
plt.tight_layout()
plt.savefig("heartbeats_combined.png")
```
