```python
import numpy as np
```

```python
# load numpy data  
path = './data_predict/'
file1 = 'ecg_4canales_1.npy'

```

```python
data1 = np.load(path + file1)
print(data1)
print(data1.shape)
print(min(data1[3]))
print(max(data1[3]))
```

```python
# loop, all 4 columns min, max values  
for i in range(data1.shape[1]):
    col_min = data1[:, i].min()
    col_max = data1[:, i].max()
    print(f"Channel {i}: Min = {col_min}, Max = {col_max}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup time axis
fs = 250  # Sampling frequency
time = np.arange(len(data1)) / fs
lead_names = ['I', 'II', 'V2', 'V5'] # custom labels 

# 2. Plot Whole Series
plt.figure(figsize=(15, 5))
plt.plot(time, data1)
plt.title("Whole Series - All 4 Channels (60 Seconds)")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend(lead_names, loc='upper right') # Applied labels 
plt.grid(True, alpha=0.3)
plt.show()

# 3. Plot First 5 Seconds
plt.figure(figsize=(15, 5))
seconds_to_show = 5
num_samples = seconds_to_show * fs

plt.plot(time[:num_samples], data1[:num_samples, :])
plt.title(f"First {seconds_to_show} Seconds - All 4 Channels")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend(lead_names, loc='upper right') # Applied labels 
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 3. Plot First 5 Seconds
plt.figure(figsize=(15, 10))
seconds_to_show = 1
num_samples = seconds_to_show * fs

plt.plot(time[:num_samples], data1[:num_samples, :])
plt.title(f"First {seconds_to_show} Seconds - All 4 Channels")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.legend(lead_names, loc='upper right') # Applied labels 
plt.grid(True, alpha=0.3)
plt.show()
```

```python
seconds_to_show = 1
num_samples = seconds_to_show * fs
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Distinct colors for each

# 2. Create Subplots (4 rows, 1 column)
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(15, 12), sharex=True)

for i in range(4):
    # Slice both time and data to the first 250 samples
    axes[i].plot(time[:num_samples], data1[:num_samples, i], color=colors[i])
    axes[i].set_ylabel(f"Lead {lead_names[i]}")
    axes[i].grid(True, alpha=0.3)

axes[-1].set_xlabel("Time (seconds)")
fig.suptitle(f"ECG Channels - First {seconds_to_show} Second", fontsize=16)

# Add common labels
axes[-1].set_xlabel("Time (seconds)")
fig.suptitle("ECG Channels - Individual View", fontsize=16)

plt.tight_layout() # Prevents labels from overlapping
plt.show()
```
